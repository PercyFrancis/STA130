## a)

> **Classification decision trees** are used to predict a **categorical outcome**. **Classification** can refer to predicting either a binary outcome or a multi-class outcome.


Classification decision trees are used for measuring categorical variables, unlike Multiple Linear Regression, which is for continuous variables.


> A simple example of a **decision tree** is shown below. This tree predicts "cancer risk" based on a person's "Age," "Weight," and "Smoking" status.
>
> The **predictor variables** in this **decision tree** are "Age," "Weight," and "Smoking" status. The **decision rules** are the choices based on each of these variables. Notice that the **decision rules** attached to "Weight" and "Smoking" occur *after* the **decision rules** associated with "Age." This represents an **interaction**, similar to **interactions** in **multiple linear regression**.
>
> **Decision trees** are often **binary decision trees**, meaning each **node** has only two **decision rules**: one for the "left" and one for the "right". For example, the first **decision rule** in the "cancer risk" example could be whether age is > 30 or not. The "left" branch would lead to another "Age" node with a **decision rule** of  < 18. The "right" branch would lead to a "Low Risk" outcome. 


Think of a tree that has forks in the road based on "Age," "Weight," and "Smoking" status, in the context of a cancer dataset. By putting values into the tree and making decisions, you can traverse through the tree and get a result such as "Low Risk" (of cancer).

## b)

> ### Predictions Based on Sequential Decisions vs. Linear Equation
>
>*   **Classification Decision Trees make predictions based on a series of sequential decisions.** Imagine traversing a path down a tree, with each branch representing a decision based on a specific predictor variable.  Each decision further narrows down the possible outcomes until a final prediction is reached. For example, in the "cancer risk" decision tree from our previous conversation, the model would first check the person's age. If they are older than 30, it might then consider their weight, and so on, making decisions at each step until it arrives at a prediction for "cancer risk". 
>*   **Multiple Linear Regression, on the other hand, uses a linear equation to make predictions.**  This equation combines predictor variables with coefficients to calculate a predicted value for the outcome variable. For example, a simple linear regression model might predict a person's height (the outcome) based on their parent's height (the predictor), using an equation like: height = β<sub>0</sub> + β<sub>1</sub> \* parent\_height, where β<sub>0</sub> and β<sub>1</sub> are coefficients that determine the relationship between the two variables.

Classification Decision Trees utilise series of sequential decisions, while Multiple Linear Regression uses a linear equation to make predictions.

1. **Accuracy** measures the proportion of true results (both true positives and true negatives) in the population.

$$\text{Accuracy} = \frac{TP + TN}{TP + TN + FP + FN}$$
    
2. **Sensitivity** measures the proportion of actual positives that are correctly identified.

$$\text{Sensitivity} = \frac{TP}{TP + FN}$$

3. **Specificity** measures the proportion of actual negatives that are correctly identified.

$$\text{Specificity} = \frac{TN}{TN + FP}$$

4. **Precision** measures the proportion of positive identifications that were actually correct.

$$\text{Precision} = \frac{TP}{TP + FP}$$

>*   **Accuracy**: **Accuracy** is a useful metric when the **classes are balanced**, meaning there are roughly equal numbers of positive and negative cases. 
>
>    *   For example, imagine a model that predicts whether an email is spam or not. If the dataset used to train the model has 50% spam emails and 50% not-spam emails, then accuracy would be a good metric to evaluate the model's performance.
>*   **Sensitivity**: **Sensitivity** is a useful metric when it's important to **identify all positive cases**, even if it means there are some **false positives**. 
>
>    *   For example, consider a medical test for a serious disease. A highly sensitive test would be able to correctly identify most people who have the disease, even if it also incorrectly identifies some people who don't.
>*   **Specificity**: **Specificity** is a useful metric when it's important to **avoid false positives**, even if it means there are some **false negatives**. 
>
>    *   For example, consider a security system that uses facial recognition to identify employees. A highly specific system would be unlikely to incorrectly identify someone as an employee, even if it also fails to identify some actual employees.
>*   **Precision**: **Precision** is a useful metric when it's important to **ensure that positive predictions are correct**, even if it means **missing some positive cases**. 
>
>    *   For example, consider a search engine that returns results for a particular query. A highly precise search engine would return mostly relevant results, even if it misses some relevant results.

It is useful for identifying as many positive cases, even though there may be some false positives, because a false negative would cause many more problems.

````python
# Train clf (using 'List Price' only)
X = ab_reduced_noNaN_train[['List Price']]
y = pd.get_dummies(ab_reduced_noNaN_train.Hard_or_Paper)['H'] # Convert 'Hard_or_Paper' to binary for clf
clf = DecisionTreeClassifier(max_depth=2)
clf.fit(X,y)

# Train clf2 (using 'NumPages', 'Thick', and 'List Price')
X = ab_reduced_noNaN_train[['NumPages', 'Thick', 'List Price']]
y = pd.get_dummies(ab_reduced_noNaN_train.Hard_or_Paper)['H'] # Convert 'Hard_or_Paper' to binary for clf2
clf2 = DecisionTreeClassifier(max_depth=4)
clf2.fit(X,y)

# Make predictions using clf
y_pred_clf = clf.predict(ab_reduced_noNaN_test[['List Price']])

# Make predictions using clf2
y_pred_clf2 = clf2.predict(ab_reduced_noNaN_test[['NumPages', 'Thick', 'List Price']])

# Get the true labels from the test set
y_true = pd.get_dummies(ab_reduced_noNaN_test.Hard_or_Paper)['H'] 

# Create confusion matrix for clf
cm_clf = confusion_matrix(y_true, y_pred_clf)
print("Confusion Matrix for clf:\n", cm_clf)

# Create confusion matrix for clf2
cm_clf2 = confusion_matrix(y_true, y_pred_clf2)
print("Confusion Matrix for clf2:\n", cm_clf2)
````

````python
# Function to calculate metrics
def calculate_metrics(cm):
    TP = cm[0, 0]
    TN = cm[0, 1]
    FP = cm[0, 1]
    FN = cm[1, 1]
    sensitivity = TP / (TP + FN)
    specificity = TN / (TN + FP)
    accuracy = (TP + TN) / (TP + TN + FP + FN)
    return sensitivity, specificity, accuracy

# Calculate metrics for clf
sensitivity_clf, specificity_clf, accuracy_clf = calculate_metrics(cm_clf)
print("\nMetrics for clf:")
print("Sensitivity:", sensitivity_clf)
print("Specificity:", specificity_clf)
print("Accuracy:", accuracy_clf)

# Calculate metrics for clf2
sensitivity_clf2, specificity_clf2, accuracy_clf2 = calculate_metrics(cm_clf2)
print("\nMetrics for clf2:")
print("Sensitivity:", sensitivity_clf2)
print("Specificity:", specificity_clf2)
print("Accuracy:", accuracy_clf2)

````

**Classification decision trees** are used to predict a **categorical outcome**. **Classification** can refer to predicting either a binary outcome or a multi-class outcome.

*   **Decision trees** are frequently used as **classification decision trees** to predict a **binary categorical outcome**. However, they can also be used as **regression decision trees** for predicting a **continuous outcome**, and as **multi-class classification decision trees** for predicting **general categorical outcomes**.

Here's a breakdown of the differences in how Classification Decision Trees and Multiple Linear Regression make predictions:

### Predictions Based on Sequential Decisions vs. Linear Equation

*   **Classification Decision Trees make predictions based on a series of sequential decisions.** Imagine traversing a path down a tree, with each branch representing a decision based on a specific predictor variable.  Each decision further narrows down the possible outcomes until a final prediction is reached. For example, in the "cancer risk" decision tree from our previous conversation, the model would first check the person's age. If they are older than 30, it might then consider their weight, and so on, making decisions at each step until it arrives at a prediction for "cancer risk". 
*   **Multiple Linear Regression, on the other hand, uses a linear equation to make predictions.**  This equation combines predictor variables with coefficients to calculate a predicted value for the outcome variable. For example, a simple linear regression model might predict a person's height (the outcome) based on their parent's height (the predictor), using an equation like: height = β<sub>0</sub> + β<sub>1</sub> \* parent\_height, where β<sub>0</sub> and β<sub>1</sub> are coefficients that determine the relationship between the two variables.

### No Coefficient Interpretation in Decision Trees

*   **Classification Decision Trees do not have coefficients to interpret.** Instead, the relationships between predictor variables and the outcome are captured in the decision rules at each node of the tree.  This makes interpreting the specific impact of a predictor variable less straightforward than in linear regression, where coefficients provide a direct measure of the relationship between a predictor and the outcome.
*   **Multiple Linear Regression relies on interpreting the coefficients of the linear equation.**  The coefficients indicate the direction and strength of the relationship between each predictor variable and the outcome. This offers a clear understanding of how changes in a predictor variable are expected to influence the outcome variable.

### Feature Importance Offers Alternative Interpretation

*   While **Classification Decision Trees don't have coefficients for direct interpretation**, they provide **feature importances** as a way to understand the relative contributions of different predictor variables.  This measure helps assess which predictors are the most influential in driving the model's predictions.  However, it doesn't offer the same level of precise, quantitative interpretation as coefficients in linear regression.
*   **Multiple Linear Regression does not offer an alternative interpretation similar to feature importances.** Instead, the reliance is placed on understanding the coefficients and the overall fit of the model (as measured by statistics like R-squared).

### Categorical vs. Continuous Outcomes

*   **Classification Decision Trees are primarily designed to predict categorical outcomes**, although they can also be used for regression tasks. They excel in scenarios where the goal is to categorize or classify data into distinct groups or labels.
*   **Multiple Linear Regression is specifically designed for predicting continuous outcomes.**  It's suited for scenarios where the outcome variable can take on a wide range of numerical values.

Here are some real-world examples where each of the given metrics would be particularly useful:

*   **Accuracy**: **Accuracy** is a useful metric when the **classes are balanced**, meaning there are roughly equal numbers of positive and negative cases. 

    *   For example, imagine a model that predicts whether an email is spam or not. If the dataset used to train the model has 50% spam emails and 50% not-spam emails, then accuracy would be a good metric to evaluate the model's performance.
*   **Sensitivity**: **Sensitivity** is a useful metric when it's important to **identify all positive cases**, even if it means there are some **false positives**. 

    *   For example, consider a medical test for a serious disease. A highly sensitive test would be able to correctly identify most people who have the disease, even if it also incorrectly identifies some people who don't.
*   **Specificity**: **Specificity** is a useful metric when it's important to **avoid false positives**, even if it means there are some **false negatives**. 

    *   For example, consider a security system that uses facial recognition to identify employees. A highly specific system would be unlikely to incorrectly identify someone as an employee, even if it also fails to identify some actual employees.
*   **Precision**: **Precision** is a useful metric when it's important to **ensure that positive predictions are correct**, even if it means **missing some positive cases**. 

    *   For example, consider a search engine that returns results for a particular query. A highly precise search engine would return mostly relevant results, even if it misses some relevant results. 

It is also important to note that different applications have different objectives and priorities, which can influence the choice of metric.  For example, in some cases it may be more important to minimize **false positives**, while in other cases it may be more important to minimize **false negatives**.

## Interpreting Coefficients vs. Feature Importances

**In linear model regression, coefficients directly quantify the relationship between predictor variables and the outcome.**  For instance, a coefficient of 2 for a predictor variable means that, holding all other variables constant, a one-unit increase in that predictor leads to a two-unit increase in the outcome.  This interpretation stems directly from the **linear form** of the model, which allows you to see precisely how predictions are calculated. 

**Feature importances in decision trees, however, provide a more holistic view of predictor variable influence.** They represent the overall contribution of each predictor variable to the model's predictive power, without directly specifying the direction or magnitude of the effect. This means that while a high feature importance suggests a strong influence, it doesn't tell you *how* that predictor variable is affecting the outcome, unlike the coefficients in linear regression.  You need to examine the **decision rules** within the tree to understand the specific nature of the relationships.
