# 1.

1. For the first research question I intend to study the relationship between loneliness and mental health. 
- My hypothesis is that loneliness leads to poor mental health.
- I found this to be interesting as it is usually taken as a fact that loneliness leads to poor mental health, but even though the hypothesis seems obvious, I still want to know if this is true and if so, how strongly the variables correlate with each other.

The population parameters that I would like to estimate are loneliness and mental health of the general population (of Canada).

2. The two variables are:
- Independent: Loneliness (None of the time (0 days) to All of the time (5-7 days))
- Dependent: Mental health (Poor to Excellent)

These variables have been chosen as they are in my hypothesis, I assumed they would relate to each other.

A bar chart would be the most effective to show this data. Specifically the bar chart will contain counts of people categorized by both their mental health and loneliness. This is useful as it shows both variables and how closely they correlate with each other.

3. I may want to bootstrap this data to find out how representative it is compared to the general population. I would bootstrap and construct confidence intervals for each category of loneliness. I will also get p-values for this analysis. The null hypothesis in this case would be:

Null Hypothesis (H₀): There is no difference in the average self-rated mental health scores between groups with different levels of loneliness in the population.

4. As stated in (1.), my hypothesis is that loneliness leads to poor mental health. This would answer my research question, the relationship between loneliness and mental health, as it would show a trend between these two variables. 

I did an analysis on this data, and the non-bootstrapped results are in favour of my hypothesis, loneliness leads to poor mental health. See below.


```python
import plotly.graph_objects as go
import plotly.subplots as sp
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.offline as pyo
df = pd.read_csv("https://raw.githubusercontent.com/pointOfive/stat130chat130/refs/heads/main/CP/CSCS_data_anon.csv")

fig = sp.make_subplots(rows=1, cols=1, subplot_titles=("LONELY_direct vs WELLNESS_self_rated_mental_health", "2", "3", "4"))
fig.add_trace(go.Histogram(x=df[df['LONELY_direct'] == 'None of the time (e.g., 0 days)']['WELLNESS_self_rated_mental_health'], name = 'None of the time'), row=1, col=1)
fig.add_trace(go.Histogram(x=df[df['LONELY_direct'] == 'Rarely (e.g. less than 1 day)']['WELLNESS_self_rated_mental_health'], name = 'Rarely (e.g. less than 1 day)'), row=1, col=1)
fig.add_trace(go.Histogram(x=df[df['LONELY_direct'] == 'Some or a little of the time (e.g. 1-2 days)']['WELLNESS_self_rated_mental_health'], name = 'Some or a little of the time (e.g. 1-2 days)'), row=1, col=1)
fig.add_trace(go.Histogram(x=df[df['LONELY_direct'] == 'Occasionally or a moderate amount of time (e.g. 3-4 days)']['WELLNESS_self_rated_mental_health'], name = 'Occasionally or a moderate amount of time (e.g. 3-4 days)'), row=1, col=1)
fig.add_trace(go.Histogram(x=df[df['LONELY_direct'] == 'All of the time (e.g. 5-7 days)]']['WELLNESS_self_rated_mental_health'], name = 'All of the time (e.g. 5-7 days)'), row=1, col=1)

fig.update_xaxes(categoryorder='array', categoryarray=['Excellent', 'Very good', 'Good', 'Fair', 'Poor'])

fig.show(renderer='png') # looks worse as png
```
