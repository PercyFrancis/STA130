1. Linear regression is used to show the relationship between two variables, predictor (independent, $x_i$) and outcome (dependent, $Y_i$) variables. Outcome varibles are the result of the predictor variable's value.
2. Slope $\beta_1$ is the rate the outcome variable changes in terms of the predictor variable.
3. The intercept coefficient $\beta_0$ is the point where the line crosses the y-axis, when the predictor variable is 0.
4. The error term, $\epsilon_i$, is the variability of the outcome variable not explained by is linear relationship with the predictor variable.

## Below is a simulation of simple linear regression

````python
import pandas as pd
import numpy as np
from scipy import stats
import statsmodels.formula.api as smf
import plotly.express as px

# Set random seed for reproducibility
np.random.seed(42)

# Define model parameters
n = 100
beta0 = 2
beta1 = 50 # 1.5
sigma = 5

# Generate predictor variable
x = stats.uniform.rvs(size=n)

# Generate error term
errors = stats.norm.rvs(loc=0, scale=sigma, size=n)

# Calculate outcome variable
Y = beta0 + beta1 * x + errors

# Create DataFrame
df = pd.DataFrame({'x': x, 'Y': Y})

# Create a scatter plot of the simulated data
fig = px.scatter(x=x, y=Y, title="Simulated Data with Theoretical Line")

# Add the theoretical line to the plot
fig.add_shape(type="line", x0=min(x), y0=beta0 + beta1 * min(x),
              x1=max(x), y1=beta0 + beta1 * max(x),
              line=dict(color="red", width=2), name="Theoretical Line")
fig.show(renderer="png")
````

1. X and Y are the independent, $x_i$, and dependent, $Y_i$, variables respectively.
2. Slope, $\beta_1$, is the value that the line on the diagram increasing. It is around 47 in this example.
3. Y-intercept, $\beta_0$, is shown to be around 3. It is where the trendline intersects the y-axis.
4. Error, $\epsilon_i$, is the spread of the values away from the trendline.

The main goal of ols is to minimize the distance of square residuals from the fitted simple linear regression line. A residual ($\text{e}_i = \hat \epsilon_i = y_i - \hat y_i$) is the vertical distance between a point and the fitted line.

Squares are used as they remove negative values and emphasises values distant from the fitted line.

````python
import scipy.stats as stats
import numpy as np
import pandas as pd
import statsmodels.formula.api as smf
import plotly.express as px

n,x_min,x_range,beta0,beta1,sigma = 20,5,5,2,3,5
x = stats.uniform(x_min, x_range).rvs(size=n)
errors = stats.norm(loc=0, scale=sigma).rvs(size=n)
Y = beta0 + beta1 * x + errors

df = pd.DataFrame({'x': x, 'y': Y})
model_data_specification = smf.ols("y~x", data=df) 
fitted_model = model_data_specification.fit() 

df['Data'] = 'Data' # hack to add data to legend 
fig = px.scatter(df, x='x',  y='y', color='Data', 
                 trendline='ols', title='Y vs. x')

# This is what `trendline='ols'` is
fig.add_scatter(x=df['x'], y=fitted_model.fittedvalues,
                line=dict(color='blue'), name="trendline='ols'")

x_range = np.array([df['x'].min(), df['x'].max()])
y_line = beta0 + beta1 * x_range
fig.add_scatter(x=x_range, y=y_line, mode='lines',
                name=str(beta0)+' + '+str(beta1)+' * x', 
                line=dict(dash='dot', color='orange'))

# Add vertical lines for residuals
for i in range(len(df)):
    fig.add_scatter(x=[df['x'][i], df['x'][i]],
                    y=[fitted_model.fittedvalues[i], df['y'][i]],
                    mode='lines',
                    line=dict(color='red', dash='dash'),
                    showlegend=False)
    
# Add horizontal line at y-bar
fig.add_scatter(x=x_range, y=[df['y'].mean()]*2, mode='lines',
                line=dict(color='black', dash='dot'), name='y-bar')

fig.show(renderer="png") # USE `fig.show(renderer="png")` FOR ALL GitHub and MarkUs SUBMISSIONS
````

### **Theoretical vs. Fitted Models**

The most important difference between a **theoretical model** and a **fitted model** is that a **theoretical model** is defined *before* observing any data, while a **fitted model** is derived *from* observed data.

*   A **theoretical model** is like a hypothesis about how variables are related. For example, the theoretical model you created in our previous conversation, $Y_i = 2 + 3x_i + \epsilon_i$, where $\epsilon_i \sim \mathcal{N}(0, \sigma)$, suggests that there is a linear relationship between the predictor variable `x` and the outcome variable `Y`, with a slope of 3 and an intercept of 2. This model also incorporates random noise, represented by  $\epsilon_i$, which accounts for the fact that the observed data will not perfectly follow the line. The sources refer to this as a **simple linear regression** model.
*   A **fitted model** is created by using statistical methods to estimate the parameters of a model based on observed data. In the context of **simple linear regression**, the **ordinary least squares (OLS)** method is commonly used.  OLS finds the line that minimizes the sum of the squared **residuals**, where the **residuals** are the differences between the observed outcome values ($Y_i$) and the predicted outcome values ($\hat{y}_i$) on the fitted line.  The sources discuss the `statsmodels` package, which can be used to fit a simple linear regression model in Python. Specifically, the  `smf.ols('Y ~ x', data=df).fit()`  function estimates the model's **intercept** ($\hat\beta_0$) and **slope** ($\hat\beta_1$) coefficients based on the data. The fitted model equation has the form $\hat y_i = \hat \beta_0 + \hat \beta_1 x_i$.

**In essence, a theoretical model represents the true (but often unknown) relationship between variables, while a fitted model is an approximation of that relationship based on the data.**  The fitted model aims to reflect the theoretical model as closely as possible, but it's essential to remember that the fitted model is based on a limited sample of data, and therefore, it is subject to **sampling variability**. This means that the fitted model parameters may not perfectly match the true parameters of the theoretical model.

The sources frequently emphasize this distinction between **theoretical** and **fitted** models, particularly within the context of **simulation** and **hypothesis testing**. The sources note that understanding **sampling distributions**, which describe the variability of sample statistics across different samples, is crucial for understanding the relationship between theoretical and fitted models.  In the context of simple linear regression, simulation can be used to explore the **sampling distribution of the slope coefficient**, helping you understand how the estimated slope coefficient might vary across different samples.

In summary, the most important difference between a fitted and theoretical model is that:

*   **Theoretical models** are proposed before data collection and represent a hypothesized relationship between variables.
*   **Fitted models** are derived from observed data and use statistical methods to estimate model parameters, providing an approximation of the theoretical relationship.

Understanding this distinction is essential for correctly interpreting statistical results and drawing meaningful conclusions from data analysis.

### Ordinary Least Squares and Simple Linear Regression

**Ordinary least squares (OLS)** is the method used to fit linear regression models to data in the provided sources. To specify a **simple linear regression** model, you can use the "formula" "y ~ x" with the `smf.ols()` function in Python. This formula corresponds to the theoretical model:

  $$Y_i \sim \mathcal{N}(\beta_0 + \beta_1 x_i , \sigma) \quad \text{ or } \quad  Y_i = \beta_0 + \beta_1 x_i + \epsilon_i, \quad \text{ where } \quad \epsilon_i \sim \mathcal{N}(0, \sigma)$$ 

where:

*   $Y_i$ is the outcome or dependent variable.
*   $x_i$ is the predictor or independent variable.
*   $\beta_0$ is the intercept coefficient.
*   $\beta_1$ is the slope coefficient.
*   $\epsilon_i$ is the error term, which is assumed to be normally distributed with mean 0 and standard deviation $\sigma$.

The `smf.ols()` function estimates the model, producing a fitted model with estimated coefficients, represented with "hats": $\hat y_i = \hat \beta_0 + \hat \beta_1 x_i$. This fitted model represents the "straight line" that best reflects the linear association observed in the dataset.

#### OLS and Residuals

**OLS** aims to minimize the sum of the squared residuals. **Residuals** ($\text{e}_i = \hat \epsilon_i = y_i - \hat y_i$) represent the vertical distance between the observed data points ($y_i$) and the fitted line ($\hat y_i$). By minimizing the sum of squared residuals, OLS finds the line that best fits the data, ensuring the smallest overall distance between the data points and the line.

However, they do note that the fitted model can be visually represented using `px.scatter(..., trendline="ols")` because "ols" indicates the same **ordinary least squares** fit of the data.

### Squaring Residuals in Ordinary Least Squares (OLS)

The sources do not explicitly address why OLS uses squared residuals instead of just residuals. However, they provide information about the purpose and mechanics of OLS, which can offer some insights:

*   **Minimizing Distance:** The primary goal of OLS is to find the line that best fits the data by minimizing the overall distance between the observed data points and the fitted line. This is achieved by minimizing the sum of the squared residuals. 

*   **Addressing Positive and Negative Deviations:** If only residuals were used, positive and negative deviations from the fitted line could cancel each other out, leading to a misleading impression of a good fit when there might be significant variations present. Squaring the residuals ensures all deviations are positive, providing a more accurate measure of the overall discrepancy between the data and the fitted line.

*   **Emphasis on Larger Errors:**  Squaring the residuals places more emphasis on larger errors, as their squared values contribute more significantly to the sum of squared residuals. This can be beneficial in identifying outliers or areas where the model's predictions are particularly inaccurate.

*   **Mathematical Convenience:** Using squared residuals simplifies the mathematical calculations involved in finding the best-fit line. It allows for the use of calculus and linear algebra techniques to derive closed-form solutions for the intercept and slope coefficient estimates. 

*   **Connection to R-squared:** The sum of squared residuals relates directly to the concept of R-squared, a measure of the proportion of variation in the outcome variable explained by the model. A smaller sum of squared residuals generally corresponds to a higher R-squared value, indicating a better fit.

While the sources do not provide a definitive answer to the "why" of squaring residuals, the points above, drawn from the provided context, suggest that this approach offers several advantages in terms of accuracy, mathematical practicality, and its connection to other key concepts in linear regression.
